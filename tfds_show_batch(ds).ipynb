```
import tensorflow as tf
import matplotlib.pyplot as plt

tf.enable_eager_execution()
```

```
def show_batch(ds):
  
  for data in ds.take(1):
    X, y = data['image'].numpy(), data['label'].numpy()
    
  rows = int(X.shape[0]**0.5)
  
  fig = plt.figure(figsize=(3*rows, 3*rows))
  fig.subplots_adjust(hspace=0.4)

  for i in range(rows*rows):
      fig.add_subplot(rows, rows, i+1)
      img = X[i]
      if img.shape[2] == 1:
        img = img.reshape(img.shape[:2])
      plt.imshow(img, cmap='gray')
      plt.xlabel("label = {}".format(y[i]))

  plt.show()
```

### mnist

```
import tensorflow_datasets as tfds

ds = tfds.load('mnist', split=tfds.Split.TRAIN)
batch_ds = ds.batch(16, drop_remainder=True)

show_batch(batch_ds)
```

### cycle_gan/apple2orange

```
import tensorflow_datasets as tfds

ds = tfds.load('cycle_gan/apple2orange', split='trainA')
batch_ds = ds.batch(16, drop_remainder=True)

show_batch(batch_ds)
```
